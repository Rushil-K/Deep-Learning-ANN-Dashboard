#### Data Preprocessing

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.model_selection import train_test_split

# Assuming nmrk2627_df is your DataFrame

# No missing values to handle as per the provided data information.

# Encoding the 'Gender' column using Label Encoding.
nmrk2627_encoder = LabelEncoder()  # Creating a LabelEncoder object for encoding.
nmrk2627_df['nmrk2627_encoded_Gender'] = nmrk2627_encoder.fit_transform(nmrk2627_df['Gender']) # Applying Label Encoding to the 'Gender' column.
nmrk2627_df = nmrk2627_df.drop('Gender', axis=1) # Removing the original 'Gender' column after encoding.


# Scaling numerical features using StandardScaler.
nmrk2627_scaler = StandardScaler() # Creating a StandardScaler object for scaling.
nmrk2627_numerical_features = ['CustomerID', 'Age', 'Income', 'Purchases', 'Clicks', 'Spent', 'nmrk2627_encoded_Gender']  # List of numerical features to be scaled.
nmrk2627_df[nmrk2627_numerical_features] = nmrk2627_scaler.fit_transform(nmrk2627_df[nmrk2627_numerical_features]) # Applying StandardScaler to the selected numerical features.

# Splitting the data into training and testing sets.
nmrk2627_X = nmrk2627_df.drop('Converted', axis=1) # Features (excluding the target variable 'Converted').
nmrk2627_y = nmrk2627_df['Converted']  # Target variable.
nmrk2627_X_train, nmrk2627_X_test, nmrk2627_y_train, nmrk2627_y_test = train_test_split(
    nmrk2627_X, nmrk2627_y, test_size=0.2, random_state=552627 # Splitting with 80% for training and 20% for testing, using a fixed random state for reproducibility.
)

# The data is now preprocessed, scaled, and split for deep learning.
# nmrk2627_X_train, nmrk2627_X_test, nmrk2627_y_train, nmrk2627_y_test are the training and testing sets.
```

```python
nmrk2627_df_processed = pd.concat([nmrk2627_X, nmrk2627_y], axis=1)
nmrk2627_df_processed.head() # Display the first few rows
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Create box plots for numerical features
nmrk2627_numerical_features = ['CustomerID', 'Age', 'Income', 'Purchases', 'Clicks', 'Spent', 'nmrk2627_encoded_Gender']
for feature in nmrk2627_numerical_features:
    plt.figure(figsize=(6, 4))
    sns.boxplot(x=nmrk2627_df_processed[feature])
    plt.title(f'Box Plot of {feature}')
    plt.show()
```

```python
import pandas as pd
import tensorflow as tf
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

# Load the preprocessed data
df = pd.read_csv("nmrk2627_df_processed.csv")

# Separate features (X) and target (y)
X = df.drop(columns=["Converted"])
y = df["Converted"]

# Split data into training & testing sets
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, random_state=552627
)

# Define the ANN model
model = tf.keras.models.Sequential([
    tf.keras.layers.Dense(64, activation="relu", input_shape=(X_train.shape[1],)),
    tf.keras.layers.Dropout(0.2),
    tf.keras.layers.Dense(32, activation="relu"),
    tf.keras.layers.Dropout(0.2),
    tf.keras.layers.Dense(1, activation="sigmoid")
])

# Compile the model
model.compile(optimizer="adam", loss="binary_crossentropy", metrics=["accuracy"])

# Train the model
print("\n🔄 Training Model...\n")
history = model.fit(X_train, y_train, epochs=10, batch_size=32, verbose=1)

# Predict on test data
y_pred_prob = model.predict(X_test)
y_pred = (y_pred_prob > 0.5).astype(int)

# Evaluate model performance
test_accuracy = accuracy_score(y_test, y_pred)
loss, acc = model.evaluate(X_test, y_test, verbose=0)

# Display results
print("\n📊 Model Evaluation:")
print(f"✅ Test Accuracy: {test_accuracy:.4f}")
print(f"📉 Test Loss: {loss:.4f}")
print(f"🎯 Final Accuracy: {acc:.4f}")
```
